```python
import random

cards = ["Ace",2,3,4,5,6,7,8,9,10,"Jack","Queen","King"]

pot = 100
play ="y"

def score(x):
    tot = 0
    for num in x:
        if num == "Ace":
            tot += 11
        elif num =="Jack":
            tot += 10
        elif num =="Queen":
            tot += 10
        elif num =="King":
            tot += 10
        elif num >=0:
            tot += num
        elif tot > 21 and "Ace" in x:
            tot -= 10
    return (tot)

print ("You have £",pot)

while play =="y":
      
    status = "Twist"
    try:
        bet = int(input("How much would you like to bet?"))
        if bet <= pot:
            print ("OK, your bet is £", bet)
            pot -= bet
        else:
            print ("You don't have enough money left")
            print ("You have £",pot,"remaining")
            continue
    except:
        print ("Please use a whole number. We don't accept stories or pennies here")
        continue
    
    user = [cards[random.randint(0,12)],cards[random.randint(0,12)]]
    dealer = [cards[random.randint(0,12)]]     
    
    while status == "Twist":
        print ("Dealer gets",dealer) 
        print ("Player has",user)

        choice = input ("Would you like to stick or twist? \nPlease enter s for stick or t for twist\n")
        if choice == "s":
            status = "Stick"
        elif choice == "t":
            user.append(cards[random.randint(0,12)])
            if score(user) > 21:
                print ("You are bust!!!")
                continue
            continue
        else:
            print ("Please choose either s for stick or t for twist\n")
    
    while score(dealer) <= 16:
        dealer.append(cards[random.randint(0,12)])
        print ("Dealer has",dealer)
        
    print ("Player has got",score(user))
    print ("Dealer has got", score(dealer))
    if score(user) > 21: 
        print ("Dealer Wins!")
        print ("You now have £", pot)
    elif score(dealer) > 21: 
        print ("The Dealer is bust!!!")
        print ("Player Wins!")
        pot += bet + bet
        print ("You now have £", pot)        
    elif score(dealer) >= score(user):
        print ("Dealer Wins!")
        print ("You now have £", pot)
    else:
        print ("Player Wins!")
        pot += bet + bet
        print ("You now have £", pot)
    
    if pot >= 500:
        print ("You have broken the bank. Go and ask daddy for a chocolate bar.")
        break
    elif pot <= 0:
        print ("You have no money left.\n GAME OVER")
        break
    
    play = input("Would you like to play again? Enter y to continue\n")
    if play != "y":
        print ("Thanks for playing!")
        break    
```
